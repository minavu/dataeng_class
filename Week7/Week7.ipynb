```python
import pandas as pd
import numpy as np

acsdf = pd.read_csv("acs2017_census_tract_data.csv", usecols=['State', 'County', 'TotalPop', 'IncomePerCap', 'Poverty'])

acsdf['TotalIncome'] = acsdf['TotalPop'] * acsdf['IncomePerCap']
acsdf['TotalPoverty'] = acsdf['TotalPop'] * acsdf['Poverty'] / 100
acsdf
```

```python
groupdf = acsdf.groupby(['State', 'County'], as_index=False).agg({'TotalPop': 'sum', 'TotalIncome': 'sum', 'TotalPoverty': 'sum'})
groupdf['TotalIncome'] = groupdf['TotalIncome'] / groupdf['TotalPop']
groupdf['TotalPoverty'] = groupdf['TotalPoverty'] / groupdf['TotalPop'] * 100
groupdf['TotalIncome'] = groupdf['TotalIncome'].round(2)
groupdf['TotalPoverty'] = groupdf['TotalPoverty'].round(2)
groupdf.rename(columns={'TotalIncome': 'IncomePerCap', 'TotalPoverty': 'Poverty'}, inplace=True)
groupdf
```

```python
County_info = groupdf
County_info.insert(0, 'ID', range(1, 1+len(County_info)))
County_info.rename(columns={'County': 'Name', 'TotalPop': 'Population', 'IncomePerCap': 'PerCapitalIncome'}, inplace=True)
County_info
```

```python
coviddf = pd.read_csv('COVID_county_data.csv', usecols=['date', 'county', 'state', 'cases', 'deaths'])

coviddf['month'] = pd.DatetimeIndex(coviddf['date']).month
coviddf['year'] = pd.DatetimeIndex(coviddf['date']).year
coviddf = coviddf.drop(columns=['date'])
coviddf
```

```python
COVID_monthly = coviddf.groupby(['year', 'month', 'state', 'county'], as_index=False).agg({'cases':'sum', 'deaths':'sum'})
COVID_monthly['county'] = COVID_monthly['county'] + ' County'
COVID_monthly.rename(columns={'year':'Year', 'month':'Month', 'state':'State', 'county':'Name', 'cases':'Cases', 'deaths':'Deaths'}, inplace=True)
COVID_monthly
```

```python
COVID_monthly = COVID_monthly.merge(County_info, how='left', on=['State', 'Name'])
COVID_monthly = COVID_monthly.drop(columns=['State', 'Name', 'Population', 'PerCapitalIncome', 'Poverty'])

COVID_monthly
```

```python
def get_id(county, state):
    loc = County_info.loc[(County_info['Name'] == county) & (County_info['State'] == state)]['ID']
    return loc.values[0]

def get_info(id, month, year):
    return COVID_monthly.loc[(COVID_monthly['Month'] == month) & (COVID_monthly['Year'] == year) & (COVID_monthly['ID'] == id)]

print(get_info(get_id('Malheur County', 'Oregon'), 2, 2021))
```

```python
print(get_info(get_id('Malheur County', 'Oregon'), 8, 2020))
```

```python
print(get_info(get_id('Malheur County', 'Oregon'), 1, 2021))
```

```python
COVID_summary = COVID_monthly.groupby(['ID'], as_index=False).agg({'Cases':'sum', 'Deaths':'sum'})
COVID_summary.rename(columns={'Cases':'TotalCases', 'Deaths':'TotalDeaths'}, inplace=True)
COVID_summary = COVID_summary.merge(County_info, how='left', on='ID').drop(columns=['State', 'Name'])
COVID_summary['TotalCasesPer100K'] = (COVID_summary['TotalCases'] / COVID_summary['Population'] * 100000).round(2)
COVID_summary['TotalDeathsPer100K'] = (COVID_summary['TotalDeaths'] / COVID_summary['Population'] * 100000).round(2)
COVID_summary
```

```python
def get_summary(id):
    return COVID_summary.loc[COVID_summary['ID'] == id]

print(get_summary(get_id('Washington County', 'Oregon')))
```
