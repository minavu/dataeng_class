```python
import pandas

df = pandas.read_csv("odot_hwy_26_crashes_2019.csv")
df
```

```python
crashdf = df[df['Record Type'] == 1]
crashdf
```

```python
vehicdf = df[df['Record Type'] == 2]
vehicdf
```

```python
partidf = df[df['Record Type'] == 3]
partidf
```

```python
crashdf = crashdf.dropna(axis='columns', how='all')
crashdf
```

```python
vehicdf = vehicdf.dropna(axis='columns', how='all')
vehicdf
```

```python
partidf = partidf.dropna(axis=1, how='all')
partidf
```

```python
df.columns.values.tolist()
```

```python
assertion = "Assertion: Every vehicle has a unique Vehicle ID"
original_list = vehicdf['Vehicle ID'].tolist()
unique_list = set(original_list)
assert len(original_list) == len(unique_list), "Every vehicle does not have a unique Vehicle ID"
print(f"{assertion} PASSED")
```

```python
assertion = "Assertion: Every crash has a unique Crash ID"
original_list = crashdf['Crash ID'].tolist()
unique_list = set(original_list)
assert len(original_list) == len(unique_list), "Every crash does not have a unique Crash ID"
print(f"{assertion} PASSED")
```

```python
assertion = "Assertion: Every crash occurred in some county"
assert all(crashdf['County Code']), "Not all crash record has county data"
print(f"{assertion} PASSED")
```

```python
assertion = "Assertion: Every crash has a DMV Crash Serial Number"
assert all(crashdf['Serial #']), "Not all crash record has a serial number"
print(f"{assertion} PASSED")
```

```python
assertion = "Assertion: Crashes are evenly distributed throughout the months of the year"
crash_month_df = crashdf.groupby(['Crash Month']).count()[['Crash ID']]
print(crash_month_df)
print(crash_month_df.describe())
std = crash_month_df.describe().loc['std', 'Crash ID']
if std < 5:
    print(f"{assertion} PASSED")
else:
    print(f"{assertion} FAILED")
print(f"The standard deviation of crashes in each month is {std}")
```

```python
assertion = "Assertion: There were thousands of crashes but not millions"
assert len(crashdf) < 1000000, "There were at least a million crashes"
assert len(crashdf) < 10000, "There were tens of thousands of crashes"
print(f"{assertion} PASSED")
```

```python
assertion = "Assertion: Every vehicle listed in the crash data was part of a known crash"
crash_ids = crashdf['Crash ID'].tolist()
for i in range(0, len(vehicdf)):
    if not vehicdf.iloc[i]['Crash ID'] in crash_ids:
        print(f"{assertion} FAILED")
        break
    else:
        print(f"{assertion} PASSED")
        break
```

```python
assertion = "Assertion: Every crash with a latitude must have a longitude"
rows_failed_assert = []
for i in range(0, len(crashdf)):
    if not crashdf.iloc[i]['Latitude Degrees'] and crashdf.iloc[i]['Longitude Degrees']:
        rows_failed_assert.append(crashdf.iloc[i])
if len(rows_failed_assert) == 0:
    print(f"{assertion} PASSED")
else:
    print(f"{assertion} FAILED")
```

```python
assertion = "Assertion: Every crash occurred during year 2019"
crash_in_year = crashdf['Crash Year'] == 2019
count = 0
for n in crash_in_year:
    if n == False:
        count += 1
if count == 0:
    print(f"{assertion} PASSED")
else:
    print(f"{assertion} FAILED")
```

```python
assertion = "Assertion: Every crash occurred on a date"
crash_has_month = crashdf['Crash Month'].tolist()
crash_has_day = crashdf['Crash Day'].tolist()
crash_has_year = crashdf['Crash Year'].tolist()
if all(crash_has_month) and all(crash_has_day) and all(crash_has_year):
    print(f"{assertion} PASSED")
else:
    print(f"{assertion} FAILED")
```

```python
crashdf.to_csv("odot_hwy_26_crashes_2019_crashdf.csv", index=False)
vehicdf.to_csv("odot_hwy_26_crashes_2019_vehicdf.csv", index=False)
partidf.to_csv("odot_hwy_26_crashes_2019_partidf.csv", index=False)
```
